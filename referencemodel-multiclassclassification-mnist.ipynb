```python
#import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Import dataset
from sklearn.datasets import fetch_openml

mnist= fetch_openml('mnist_784', as_frame=False) 
```

```python
X, Y = mnist.data, mnist.target

print(X)
print(X.shape)

print(Y)
print(Y.shape)
```

# **Visuzlize data**

```python
#define a function to plot digits

def plot_digit(image_data,label=None):
    image = image_data.reshape(28,28)
    plt.imshow(image, cmap="binary")
    plt.axis("off")
    plt.title(label)
    plt.show()
    

#show random digit digit
random_number = np.random.randint(low=0, high=1000)
plot_digit(X[random_number],Y[random_number])
```

**no pre-processing of the data is required**

# Train Test Split

```python
#creating training and test split

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.1)
```

# Train Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

forest_clf = RandomForestClassifier(random_state=42)
forest_clf.fit(X_train, Y_train)
```

**Check prediction on a random input**

```python
#predict on some random number
random_number = np.random.randint(low=0, high=10000)
predicted = forest_clf.predict(X_train[random_number:random_number+1,])
print("Predicted : {}, Actual: {}".format(predicted, Y_train[random_number]))
```

**Check cross validation accurcy**

```python
from sklearn.model_selection import cross_val_score
cv_accuracy = cross_val_score(forest_clf, X_train, Y_train, cv=3, scoring="accuracy")
print(cv_accuracy)
```

**Generate Confusion Matrix**<br>
Note: **Rows** shows **actual** and **Column** shows **Predicted**<br>
Syntax is <br>
<pre>

**Also Note**<br>
* False Positve is type 1 error
* False Negative is type 2 error

```python
from sklearn.metrics import confusion_matrix
Y_train_predicted = forest_clf.predict(X_train)
cm = confusion_matrix(Y_train,Y_train_predicted)
print(cm)

```

**Get Precision Recall and F1 Score**
* Precision is accuracy of positive predictions = TP/(TP+FP)
* Recall is also called as sensitivity or True Positive Rate, basically its out of total positive how many are identified = TP/(TP+FN)
* F1 Score is harmonic mean of precision and recall

**Confusion matrix display**

```python
#confusion matrix display
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.model_selection import cross_val_predict

Y_train_predicted = cross_val_predict(forest_clf, X_train, Y_train, cv=3)

ConfusionMatrixDisplay.from_predictions(Y_train, Y_train_predicted)
plt.title("Without Normalization")
plt.show()

ConfusionMatrixDisplay.from_predictions(Y_train, Y_train_predicted, normalize="true", values_format=".0%")
plt.title("With Normalization")
plt.show()
```
